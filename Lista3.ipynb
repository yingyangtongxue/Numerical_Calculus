```python
def print_table(k,x,e,f):
  print("--------------------------------------------------")
  print('| %2d | %.10f | %.6e | %.6e |' % (k, x, e, f))
```

1) Use o método de Newton para resolver os exemplos abaixo.
Para cada exemplo, o chute inicial deve ser calculado como
o ponto médio do intervalo dado. Use a tolerância $\epsilon$ = $10^{-3}$
para seu critério de parada. Para cada exemplo, construa
uma tabela, mostrando os valores de $x^{(k)}$, $e^{(k)}$ = $| x^{(k)}$ - $x^{(k-1)} |$, $|f(x^{(k)})|$ para cada iteração k.



a) $x^{3}-2x^{2}-5=0,\space[1,4]$

b) $x^{3}+3x^{2}-1=0,\space[-3,-2]$

c) $x-cos(x)=0,\space[0,\pi/2]$

d) $x-0.8-0.2sin(x)=0,\space[0,\pi/2]$
 

```python
def newton_method(exp,chute_inicial,itMax,tol):
  init_printing(use_unicode=True)
  x = symbols('x')
  f = lambdify(x, exp)
  df = lambdify(x, diff(exp, x))
  x = chute_inicial
  iter= 0
  x_ant = 0
  print_table(iter,x,abs(x_ant - x), abs(f(x)))
  while iter < itMax and (abs(x_ant - x) > tol or f(x) > tol):
    x_ant = x
    x = x_ant - f(x_ant)/df(x)
    e = abs(x-x_ant)
    print_table(iter+1,x,e, abs(f(x)))
    iter += 1
```

```python
print("1-a) Método de Newton:\n")
var('x')
exp = (x**3)-(2*x**2)-5
a=1
b=4
medium_point=(a+b)/2
print("| k  |     x^(k)    |    |e(k)|    |   |f(x(k))|  |")
newton_method(exp,medium_point,itMax = 100,tol=10e-3)
```

2) Repita o exercício acima utilizando o método da
bissecção.

```python
def bisseccao_method(f, a, b, tol, itmax):
    init_printing(use_unicode=True)
    x = symbols('x')
    f = lambdify(x, exp)
    i = 0  
    x_ant = 0
    x = (b+a)/2  
    e = abs(x_ant - x)
    print_table(i,x,e, abs(f(x))) 
    while (i < itmax) and (f(x) != 0) and ((b-a)/2 > tol):
        if (np.sign(f(a)) == np.sign(f(x))):  # f(a) * f(x) > 0
            a = x  
            x_ant = a
        else:  
            b = x 
        
        e = abs(x_ant - x)
        print_table(i+1,x,e, abs(f(x)))  
        x_ant = x
        x = (b+a)/2  
        i = i+1 
```

```python
var('x')
exp = (x**3)-(2*x**2)-5
a=1
b=4

print("2-a) Método da Bissecção:\n")
print("| k  |     x^(k)    |    |e(k)|    |   |f(x(k))|  |")
bisseccao_method(exp, a, b, tol=10e-3, itmax=100)
```
